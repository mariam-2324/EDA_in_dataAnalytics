# UNIVARIATE:

```python
#find a percentage for Store_Area column
df['Store_Area'].value_counts() / len(df) * 100
#round the percentage to 2 decimal places
#formula=>round(number, decimalplaces)
round(df['Store_Area'].value_counts() / len(df) * 100, 2)
```

```python
class_counts=round(df['Store_Sales'].value_counts() / len(df) * 100, 2)
#sort values in ascending order
class_counts.sort_values(ascending=False)
```

# BIVARIATE:for two variables a bivariate composition

```python
#find the values counts in percentage for store_sales
df_id = [df['Store_Sales']=='sales']
df_id["Store_Sales"].value_counts() / len(df) * 100
```

```python
#group by 2 items in data, and their count appears within different types of data set.
df.groupby(['Store_Sales']).value_counts() / len(df) * 100
df.groupby(['Store_Sales', 'Store_Area']).size() / len(df) * 100
round(df.groupby(['Store_Sales', 'Store_Area']).size() / len(df) * 100, 2).sort_values(ascending=False)
```
